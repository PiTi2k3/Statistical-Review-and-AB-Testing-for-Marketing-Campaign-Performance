```python
import pandas as pd
import numpy as np
import statsmodels.api as sm 
control=pd.read_csv("control_group.csv",sep=";")
test=pd.read_csv("test_group.csv",sep=";")
def clean_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_")
        .str.replace("#", "")
        .str.replace("[", "")
        .str.replace("]", "")
    )
    return df

control = clean_columns(control)
test = clean_columns(test)

control.columns
```

```python
#metric, Conversion Rate = Purchase / Impressions
control_cr = control["_of_purchase"].sum() / control["_of_impressions"].sum()
test_cr = test["_of_purchase"].sum() / test["_of_impressions"].sum()

print("Control CR:", control_cr)
print("Test CR:", test_cr)
```

```python
H₀: Conversion rate Control = Test

H₁: Conversion rate Control ≠ Test
```

```python
success = np.array([
    control["_of_purchase"].sum(),
    test["_of_purchase"].sum()
])

nobs = np.array([
    control["_of_impressions"].sum(),
    test["_of_impressions"].sum()
])

z_stat, p_value = sm.stats.proportions_ztest(success, nobs)

z_stat, p_value
```

```python
if p_value < 0.05:
    print("Reject H0: Test group hiệu quả hơn (có ý nghĩa thống kê)")
else:
    print("Fail to reject H0: Không đủ bằng chứng cho sự khác biệt")
```
